```python
# Create a LightGBM dataset
train_data = lgb.Dataset(X_train, label=y_train)
test_data = lgb.Dataset(X_test, label=y_test, reference=train_data)

# Define the parameter grid for grid search
param_grid = param_grid = {
    'boosting_type': ['gbdt'],
    'objective': ['regression'],
    'num_leaves': [10,30,50],
    'learning_rate': [0.01, 0.1 , 1.0],
    'n_estimators': [100, 150 , 200],
    'reg_lambda': [0.01, 0.1,1.0]
}
#Reg-lambda is a parameter that reduces overfitting and underfitting of the data

# Create a LightGBM Regressor
lgb_model = lgb.LGBMRegressor()

# Create GridSearchCV object - Hyper parameter tuning to select the best parameters for the light gbm model to reduce overfitting of the data
grid_search = GridSearchCV(estimator=lgb_model, param_grid=param_grid, cv=3, scoring='neg_mean_squared_error', verbose=1)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Print the best parameters and corresponding score
print("Best Parameters: ", grid_search.best_params_)
print("Best Score: ", grid_search.best_score_)

# Get the best model
best_lgb_model = grid_search.best_estimator_
```

```python
#Learning curves to understand the performance of the model during each interation
# Both curves stabilizing at relatively low and close values indicate good generalization without significant overfitting (since both training and testing RMSE are close).
train_results = []
test_results = []

for num_trees in range(1, 201, 10):  # Assuming you want to evaluate every 10 iterations
    best_lgb_model.set_params(n_estimators=num_trees)
    best_lgb_model.fit(X_train, y_train)
    
    train_pred = best_lgb_model.predict(X_train)
    test_pred = best_lgb_model.predict(X_test)
    
    train_rmse = mean_squared_error(y_train, train_pred, squared=False)
    test_rmse = mean_squared_error(y_test, test_pred, squared=False)
    
    train_results.append(train_rmse)
    test_results.append(test_rmse)

# Plot learning curves
plt.figure(figsize=(10, 6))
plt.plot(range(1, 201, 10), train_results, label='Training RMSE', marker='o')
plt.plot(range(1, 201, 10), test_results, label='Testing RMSE', marker='o')
plt.title('Learning Curves')
plt.xlabel('Number of Trees')
plt.ylabel('RMSE')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Cross validation technique to confirm the absence of overfitting of the data
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score
import lightgbm as lgb
import numpy as np

num_of_folds = 5
kf = KFold(n_splits=num_of_folds, shuffle=True, random_state=42)

clf = lgb.LGBMRegressor(reg_lambda=0.1)

# Assuming X_train, y_train, X_test, y_test are your training and test data

# Initialize an array to store predictions
y_test_pred_cv = np.zeros_like(y_test, dtype=float)

# Perform cross-validated predictions on the test set
for train_index, test_index in kf.split(X_test):
    X_test_fold, y_test_fold = X_test.iloc[test_index], y_test.iloc[test_index]
    clf.fit(X_train, y_train)
    y_test_pred_fold = clf.predict(X_test_fold)
    y_test_pred_cv[test_index] = y_test_pred_fold

# Calculate R-squared manually
r2_test_cv = r2_score(y_test, y_test_pred_cv)
print("Cross-validated Test Set R-squared:", r2_test_cv)
```

```python
#Defining a function to predict the 'LOS' for a pateint based on the surgeon, surgery and room type 
import pandas as pd

def suggest_3los(surgeon_name, surgery_name, pay_type):
    surgeon_name = surgeon_name.lower()
    surgery_name = surgery_name.lower()
    pay_type = pay_type.lower()

    # Replace this line with your DataFrame retrieval or creation
    df_lower = sur_df_mod.apply(lambda x: x.astype(str).str.lower())

    # Use '==' for exact matching
    filtered_data = df_lower[(df_lower['SURGEN_NAME'] == surgeon_name) &
                             (df_lower['SURGERY'] == surgery_name) &
                             (df_lower['PAY_TYPE'] == pay_type)]

    print(f"Filtered Data:\n{filtered_data}")

    if filtered_data.empty:
        # Use '==' for exact matching in similar_data
        similar_data = df_lower[(df_lower['SURGEN_NAME'] == surgeon_name) &
                                (df_lower['SURGERY'] == surgery_name)]
        print(f"Similar Data:\n{similar_data}")

        if similar_data.empty:
            return ["Not enough data for suggestion"]

        # Convert 'LOS' column to numeric
        similar_data.loc[:, 'LOS'] = pd.to_numeric(similar_data['LOS'], errors='coerce')
        mean_los = similar_data['LOS'].mean()
        std_los = similar_data['LOS'].std()

        # Consider data span for suggestions
        suggested_los = [mean_los - std_los, mean_los, mean_los + std_los]
    else:
        # Convert 'LOS' column to numeric
        filtered_data.loc[:, 'LOS'] = pd.to_numeric(filtered_data['LOS'], errors='coerce')
        # Consider data span for suggestions
        suggested_los = [filtered_data['LOS'].min(), filtered_data['LOS'].mean(), filtered_data['LOS'].max()]

    # Convert the suggested LOS values to a set and then back to a list to get unique values
    suggested_los_unique = list(set(suggested_los))

    return suggested_los_unique

# Example usage
surgeon_name = 'AHMED ALI M'  # Replace with your input
surgery_name = 'LAP CHOLE'    # Replace with your input
pay_type = 'SINGLE NON - AC'  # Replace with your input

suggested_los_values = suggest_3los(surgeon_name, surgery_name, pay_type)
formatted_suggested_los = [int(value) for value in suggested_los_values]
print(f"Suggested LOS Values: {formatted_suggested_los}")
```

```python
# Defining a function to compute the breakdown of the final bill amount
def get_bill_breakdown(total_amt_input, surgery_name, surgeon_name, pay_type, df):
    # Filter the dataset based on input values
    filtered_data1 = df[(df['SURGERY'] == surgery_name) &
                       (df['SURGEN_NAME'] == surgeon_name) &
                       (df['PAY_TYPE'] == pay_type)]

    # Check if the filtered data is empty
    if filtered_data1.empty:
        return None  # Return None indicating not enough data

    # Compute mean values for similar cases
    mean_values = filtered_data1[['CONSULTATION_AMT', 'ROOM_TOTAL', 'PHARMACY_AMT',
                                 'TEST_AMT', 'SERVICE_CHARGES',
                                 'SURGEON_CHARGES', 'ANAESTHESIST_CHARGES']].mean(skipna=True)

    breakdown = mean_values * (total_amt_input / mean_values.sum())
    return breakdown

```
